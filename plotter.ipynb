```python
from ipywidgets import FloatProgress
from IPython.display import display
import numpy as np
import matplotlib.pyplot as plt
from os import system
import subprocess, os, sys
```

### Simulation parameters:

```python
filename       = 'sim'
size           = 10
discretization = 50
iterations     = 100
individuals    = 1000
```

### Plot with killing border:

```python
wall = 0
cmd = ["./"+filename, str(size), str(discretization), str(iterations), str(individuals), str(wall)]

p = subprocess.Popen(cmd,
                     stdout=subprocess.PIPE,
                     stderr=subprocess.STDOUT)

f = FloatProgress(min=1, max=iterations)
display(f)
for line in p.stdout:
    f.value += 1
f.close()

Z_init = np.loadtxt('init_density.csv', delimiter=',')
Z = np.loadtxt('density.csv', delimiter=',')
pop = np.loadtxt('population.csv', delimiter=',')
moment1 = np.loadtxt('moment1.csv', delimiter=',')

plt.imshow(Z_init)
plt.show()
plt.imshow(Z)
plt.show()
plt.plot(pop, '-')
plt.show()
plt.plot(moment1, '-')
plt.show()
for i in range(int(iterations / 10)):
    density = np.loadtxt('density' + str(i*10 + 1) + '.csv', delimiter=',')
    plt.imshow(density)
    plt.show()
```
